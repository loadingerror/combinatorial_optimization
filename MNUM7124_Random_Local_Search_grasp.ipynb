```python
# reads TSP file and return the distance matrix
def read_tsp_file(filename):
    distance_matrix = []
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            reader = csv.reader(f)
            next(reader)  
            distance_matrix = [list(map(float, row[2:])) for row in reader]  
    else:
        print(f"File {filename} not found.")
        return None
    return distance_matrix

# reads all TSP files in a directory and return a dictionary of distance matrices
def read_all_tsp_files(directory):
    tsp_files = {}
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            full_path = os.path.join(directory, filename)
            tsp_files[filename] = read_tsp_file(full_path)
    return tsp_files

# calculate the total distance of a TSP tour
def calculate_total_distance(distance_matrix, tour):
    total_distance = 0
    for i in range(len(tour)):
        total_distance += distance_matrix[tour[i - 1]][tour[i]]
    return total_distance

# Generate the neighborhood by swapping 
def generate_neighborhood(tour):
    neighborhood = []
    for i in range(len(tour)):
        for j in range(i + 1, len(tour)):
            neighbor = tour[:]
            neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
            neighborhood.append(neighbor)
    return neighborhood

# First Improvement Strategy
def first_improvement(distance_matrix, initial_tour):
    current_tour = initial_tour[:]
    current_distance = calculate_total_distance(distance_matrix, current_tour)
    
    while True:
        neighborhood = generate_neighborhood(current_tour)
        improvement_found = False
        
        for neighbor in neighborhood:
            neighbor_distance = calculate_total_distance(distance_matrix, neighbor)
            if neighbor_distance < current_distance:
                current_tour = neighbor
                current_distance = neighbor_distance
                improvement_found = True
                break
        
        if not improvement_found:
            break
    
    return current_tour, current_distance

# Best Improvement Strategy
def best_improvement(distance_matrix, initial_tour):
    current_tour = initial_tour[:]
    current_distance = calculate_total_distance(distance_matrix, current_tour)
    
    while True:
        neighborhood = generate_neighborhood(current_tour)
        best_neighbor = None
        best_neighbor_distance = current_distance
        
        for neighbor in neighborhood:
            neighbor_distance = calculate_total_distance(distance_matrix, neighbor)
            if neighbor_distance < best_neighbor_distance:
                best_neighbor = neighbor
                best_neighbor_distance = neighbor_distance
        
        if best_neighbor_distance < current_distance:
            current_tour = best_neighbor
            current_distance = best_neighbor_distance
        else:
            break
    
    return current_tour, current_distance

# Solve TSP for each file in thefolder
def solve_tsp_for_instances(folder_path):
    results = []
    files = read_all_tsp_files(folder_path)  
    
    for filename, distance_matrix in files.items():
        if distance_matrix is not None:  
            instance_name = filename.split('.')[0]
            num_cities = len(distance_matrix)
            initial_tour = list(range(num_cities))
            random.shuffle(initial_tour)

            # first improvement application
            best_tour_fi, best_distance_fi = first_improvement(distance_matrix, initial_tour)
            results.append([instance_name, 'first_improvement', best_distance_fi, best_tour_fi])

            # best improvement application
            best_tour_bi, best_distance_bi = best_improvement(distance_matrix, initial_tour)
            results.append([instance_name, 'best_improvement', best_distance_bi, best_tour_bi])

    return results

# results to csv
def write_results_to_csv(results, output_file):
    with open(output_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['instance', 'strategy', 'best_OF', 'best_SOL'])
        writer.writerows(results)

# run
folder_path = r'C:\Users\55489\Documents\GitHub\combinatorial_optimization\TSP'  
results = solve_tsp_for_instances(folder_path)
write_results_to_csv(results, 'marina_ferreira_TSP_local.csv')
```

```python
# reads TSP file and return the distance matrix
def read_tsp_file(filename):
    distance_matrix = []
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            reader = csv.reader(f)
            next(reader)  # Skip header if necessary
            distance_matrix = [list(map(float, row[2:])) for row in reader]  
            
        # Ensure matrix is square
        num_rows = len(distance_matrix)
        for row in distance_matrix:
            if len(row) != num_rows:
                print(f"Error: Distance matrix in {filename} is not square.")
                return None
    else:
        print(f"File {filename} not found.")
        return None
    
    return np.array(distance_matrix)  

# reads ALL TSP files in a directory and return a dictionary of distance matrices
def read_all_tsp_files(directory):
    tsp_files = {}
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            full_path = os.path.join(directory, filename)
            tsp_files[filename] = read_tsp_file(full_path)
    return tsp_files

# calculating the total distance of a TSP
def calculate_total_distance(distance_matrix, tour):
    total_distance = sum(distance_matrix[tour[i - 1], tour[i]] for i in range(len(tour)))
    return total_distance

# Generate the neighborhood by swapping
def generate_neighborhood(tour, num_neighbors=100):
    neighborhood = []
    for _ in range(num_neighbors): 
        i, j = random.sample(range(len(tour)), 2) 
        neighbor = tour[:]
        neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
        neighborhood.append(neighbor)
    return neighborhood

# First Improvement
def first_improvement(distance_matrix, initial_tour):
    current_tour = initial_tour[:]
    current_distance = calculate_total_distance(distance_matrix, current_tour)
    
    while True:
        neighborhood = generate_neighborhood(current_tour)
        improvement_found = False
        
        for neighbor in neighborhood:
            neighbor_distance = calculate_total_distance(distance_matrix, neighbor)
            if neighbor_distance < current_distance:
                current_tour = neighbor
                current_distance = neighbor_distance
                improvement_found = True
                break
        
        if not improvement_found:
            break
    
    return current_tour, current_distance

# Best Improvement!
def best_improvement(distance_matrix, initial_tour):
    current_tour = initial_tour[:]
    current_distance = calculate_total_distance(distance_matrix, current_tour)
    
    while True:
        neighborhood = generate_neighborhood(current_tour)
        best_neighbor = None
        best_neighbor_distance = current_distance
        
        for neighbor in neighborhood:
            neighbor_distance = calculate_total_distance(distance_matrix, neighbor)
            if neighbor_distance < best_neighbor_distance:
                best_neighbor = neighbor
                best_neighbor_distance = neighbor_distance
        
        if best_neighbor_distance < current_distance:
            current_tour = best_neighbor
            current_distance = best_neighbor_distance
        else:
            break
    
    return current_tour, current_distance

# Greedy randomized construction 
def greedy_randomized_construction(distance_matrix, alpha=0.3):
    num_cities = len(distance_matrix)
    tour = [random.choice(range(num_cities))]  
    remaining_cities = set(range(num_cities)) - set(tour)
    
    while remaining_cities:
        last_city = tour[-1]
        distances_to_last_city = [(city, distance_matrix[last_city][city]) for city in remaining_cities]
        
        distances_to_last_city.sort(key=lambda x: x[1])
        limit = max(1, int(alpha * len(distances_to_last_city))) 
        candidates = distances_to_last_city[:limit]
        next_city = random.choice(candidates)[0]
        tour.append(next_city)
        remaining_cities.remove(next_city)
    
    return tour

# GRASP with local search - Reduced iterations and added early termination
def grasp_tsp(distance_matrix, max_iterations=10, alpha=0.3, search_strategy='first_improvement'):
    best_tour = None
    best_distance = float('inf')
    no_improvement_counter = 0

    for _ in range(max_iterations):
        #construct a random greedy solution
        initial_tour = greedy_randomized_construction(distance_matrix, alpha)

        # performing local search 
        if search_strategy == 'first_improvement':
            tour, distance = first_improvement(distance_matrix, initial_tour)
        else:
            tour, distance = best_improvement(distance_matrix, initial_tour)
        
        #keep track of the best solution found
        if distance < best_distance:
            best_tour = tour
            best_distance = distance
            no_improvement_counter = 0
        else:
            no_improvement_counter += 1

        #early termination if no improvement is found after 5 iterations
        if no_improvement_counter > 5:
            break
    
    return best_tour, best_distance

#solve TSP for each file in the folder using GRASP
def solve_tsp_for_instances(folder_path, max_iterations=10, alpha=0.3):
    results = []
    files = read_all_tsp_files(folder_path)  
    
    for filename, distance_matrix in files.items():
        if distance_matrix is not None:  
            instance_name = filename.split('.')[0]

            # GRASP first improvement
            best_tour_fi, best_distance_fi = grasp_tsp(distance_matrix, max_iterations, alpha, 'first_improvement')
            results.append([instance_name, 'GRASP_first_improvement', best_distance_fi, best_tour_fi])

            # GRASPbest improvement
            best_tour_bi, best_distance_bi = grasp_tsp(distance_matrix, max_iterations, alpha, 'best_improvement')
            results.append([instance_name, 'GRASP_best_improvement', best_distance_bi, best_tour_bi])

    return results


def write_results_to_csv(results, output_file):
    with open(output_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['instance', 'strategy', 'best_OF', 'best_SOL'])
        writer.writerows(results)

# GRASP algorithm
folder_path = r'C:\Users\55489\Documents\GitHub\combinatorial_optimization\TSP'  
results = solve_tsp_for_instances(folder_path, max_iterations=50, alpha=0.3)  # Reduced iterations to 50
write_results_to_csv(results, 'marina_ferreira_TSP_GRASP_optimized.csv')
```
